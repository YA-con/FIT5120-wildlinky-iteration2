## Extract the Victoria Local Council info

```python
import pandas as pd
```

```python
excel_path = "2025-VIC-LOCAL-GOVERNMENT-DIRECTORY-March-2025.xlsx"
df = pd.read_excel(excel_path, sheet_name='DIRECTORY', skiprows=5)
```

```python
df.head(10)
```

```python
import pandas as pd


cleaned_data = []

for index, row in df.iterrows():
    if row.isna().all():
        cleaned_data.append({})  # Add empty dict as a breakpoint
    else:
        row_dict = {
            key: value
            for key, value in row.items()
            if pd.notna(value)
        }
        cleaned_data.append(row_dict)

```

```python
print(cleaned_data)
```

```python
grouped_list = []
for item in cleaned_data:
  if item != {}:
    for key,value in item.items():
      grouped_list.append(value)
  else:
    grouped_list.append(0)


```

```python
print(grouped_list)
```

```python
count=0
item_list=[]
main_list = []
for item in grouped_list:
  if item != 0:
    item_list.append(item)
  else:
    main_list.append(item_list)
    item_list=[]
```

```python
print(main_list)
```

```python
count15=0
count17=0
print(len(main_list))
for item in main_list:
  if(len(item)==15):
    count15 +=  1
  # print(len(item))
  if(len(item)==17):
    count17 += 1
    # print(item)

print('count15:',count15)
print('count17:',count17)
```

```python
main_keys = [
    'council_name',
    'addr_line_1',
    'tel_phone',
    'ceo_name',
    'address_line_2',
    'post_code',
    'fax',
    'mayor',
    'address_line_3',
    'email',
    'website'
]

final_data = []

for item in main_list:
    if len(item) == 15:
        data = {
            'council_name': item[0],
            'addr_line_1': item[1],
            'tel_phone': item[3],
            'ceo_name': item[5],
            'address_line_2': item[6],
            'post_code': item[7],
            'fax': None,  # Not available in this format
            'mayor': item[11],
            'address_line_3': item[12],
            'email': item[9],
            'website': item[14]
        }
    elif len(item) == 17:
        data = {
            'council_name': item[0],
            'addr_line_1': item[1],
            'tel_phone': item[3],
            'ceo_name': item[5],
            'address_line_2': item[6],
            'post_code': item[7],
            'fax': item[9],
            'mayor': item[11],
            'address_line_3': item[12],
            'email': item[14],
            'website': item[16]
        }
    else:
        # You could skip or log unexpected entries
        continue

    final_data.append(data)

# Example output
for entry in final_data[:3]:
    print(entry)
```

```python
# import pandas as pd


df = pd.DataFrame(final_data)

# Combine the address parts into one string (separated by comma and space)
df['address'] = df['addr_line_1'].astype(str) + ', ' + \
                df['address_line_2'].astype(str) + ', ' + \
                df['address_line_3'].astype(str)

# Drop the original columns
df.drop(['addr_line_1', 'address_line_2', 'address_line_3'], axis=1, inplace=True)

df['post_code'] = df['post_code'].astype('Int64')  # Handles NaNs safely

# Optional: preview the DataFrame
# print(len(df))
```

```python
df.to_csv('vic_council_detais.csv', index=False)
```
